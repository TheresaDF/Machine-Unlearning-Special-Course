# Replicate Figure 3

```python
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as snb
snb.set_theme()
snb.set(font_scale=1.5)

```

### Read Data 

```python
data = np.load("data.npy", allow_pickle=True).item()
X_train = data['X_train']
```

```python
res = np.loadtxt("Results/figure_3_10900.txt")
res = res[res != 0][:600]
min_indices = np.argsort(res)[:10]
max_indices = np.argsort(res)[-10:]
```

Maximum norm 

```python
%matplotlib inline 
fig, ax = plt.subplots(1, 10, figsize = (18, 3))
for i in range(10):
    ax[i].imshow(X_train[max_indices[i]].reshape(28, 28))
    ax[i].axis("off")

plt.show()
```

Minimum norm 

```python
fig, ax = plt.subplots(1, 10, figsize = (18, 3))
for i in range(10):
    ax[i].imshow(X_train[min_indices[i]].reshape(28, 28))
    ax[i].axis("off")
plt.show()
```
