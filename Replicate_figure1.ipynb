# Test Accuracies 

```python
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as snb
snb.set_theme()
snb.set(font_scale=1.5)

```

### Figure 1 (left)

```python
# accuracies = np.loadtxt("Results/old_results/figure1_left_old.txt")
accuracies = np.loadtxt("Results/figure1_left.txt")
accuracies_extra = np.loadtxt("Results/figure1_left_extra.txt")
accuracies = (accuracies + accuracies_extra) / 2.0 
sigmas = np.logspace(-2, 2)

plt.figure(figsize = (8, 6))
plt.plot(sigmas, accuracies[:, 0], color = [0.1, 0.6, 0.2], label = "$\lambda = 10^{-4}$")
plt.plot(sigmas, accuracies[:, 1], color = [0.1, 0.75, 0.3], label = "$\lambda = 10^{-3}$")
plt.plot(sigmas, accuracies[:, 2], color = [0.1, 0.88, 0.4], label = "$\lambda = 10^{-2}$")
plt.plot(sigmas, accuracies[:, 3], color = [0.1, 1, 0.5], label = "$\lambda = 10^{-1}$")
plt.xlabel("$\sigma$")
plt.ylabel("Test accuracy")
plt.xticks([1e-2, 1e-1, 1e0, 1e1, 1e2], {"$10^{-2}$", "$10^{-1}$", "$10^{0}$", "$10^{1}$", "$10^{2}$"})
plt.yticks([0.75, 0.8, 0.85, 0.9, 0.95], {'75', '80', '85', '90', '95'})
plt.xscale('log')
plt.legend()
plt.show()
```

### Replicate figure 1 middle 

```python
accuracies = np.loadtxt("Results/figure1_mid.txt")
eps = np.logspace(-3, 4)

plt.figure(figsize = (8, 6))
plt.plot(eps, accuracies[0], color = [0.1, 0.6, 0.2], label = "$\lambda = 10^{-4}$")
plt.plot(eps, accuracies[1], color = [0.1, 0.75, 0.3], label = "$\lambda = 10^{-3}$")
plt.plot(eps, accuracies[2], color = [0.1, 0.88, 0.4], label = "$\lambda = 10^{-2}$")
plt.plot(eps, accuracies[3], color = [0.1, 1, 0.5], label = "$\lambda = 10^{-1}$")
plt.xlabel("$\epsilon$")
plt.ylabel("Test accuracy")
plt.xticks([1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3, 1e4], {"$10^{-3}$", "$10^{-2}$", "$10^{-1}$", "$10^{0}$", "$10^{1}$","$10^{2}$", "$10^{3}$", "$10^{4}$"})
plt.xscale('log')
plt.ylim([0.75, 1.0])
plt.legend()
plt.show()
```

### Replicate right plot in figure 1
